```python
import numpy as np

train_indexes = np.load("../data/annotations/train_stanford_StanfordExtra_v12.npy")
train_skip_indexes = np.load("../data/annotations/train_skip_stanford_StanfordExtra_v12.npy")

print(len(train_indexes))

val_indexes = np.load("../data/annotations/val_stanford_StanfordExtra_v12.npy")
val_skip_indexes = np.load("../data/annotations/val_skip_stanford_StanfordExtra_v12.npy")

print(len(val_indexes))

test_indexes = np.load("../data/annotations/test_stanford_StanfordExtra_v12.npy")

print(len(test_indexes))

total = len(train_indexes) + len(val_indexes) + len(test_indexes)

print(total)
print(len(train_indexes) / total)

new_train_indexes = np.copy(train_indexes)
new_val_indexes = np.array([])

print(new_train_indexes)
print(new_val_indexes)

val_skip_np = set(val_skip_indexes.flatten())

print(val_skip_np)

amount_to_move = 2359
count_moved = 0

for i, val in enumerate(val_indexes):
    should_move = (count_moved < amount_to_move) and (val not in val_skip_np)
    if (should_move):
        new_train_indexes = np.append(new_train_indexes, val)
        count_moved = count_moved  + 1
    else:
        new_val_indexes = np.append(new_val_indexes, val)


print(len(new_train_indexes))
print(len(new_val_indexes))

np.save("../data/annotations/train_stanford_StanfordExtra_v12_new_split.npy", np.array(new_train_indexes, dtype=int))
np.save("../data/annotations/val_stanford_StanfordExtra_v12_new_split.npy", np.array(new_val_indexes, dtype=int))
np.save("../data/annotations/test_stanford_StanfordExtra_v12_new_split.npy", np.array(test_indexes, dtype=int))


    

```

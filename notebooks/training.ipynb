```python
import sys
sys.path.insert(0, '..')

from src.core import TrainDataLoaderNoShuffle, ValidationDataLoader, TestDataLoader
from matplotlib import cm
import matplotlib.pyplot as plt

# PyTorch TensorBoard support
from torch.utils.tensorboard import SummaryWriter
from datetime import datetime
from matplotlib import colormaps

cm = colormaps['inferno']

def display_images(data_loader):
    # display images
    for batch_idx, inputs in enumerate(data_loader):
        images, heatmaps = inputs
        fig = plt.figure(figsize=(14, 7))
        heatmap_fig = fig = plt.figure(figsize=(14, 7))
        for i in range(8):
            ax = fig.add_subplot(2, 4, i + 1, xticks=[], yticks=[])
            plt.imshow(images[i].numpy().transpose(1, 2, 0))
        break


# display heatmaps
def display_heatmaps(data_loader):
    for batch_idx, inputs in enumerate(data_loader):
        images, heatmaps = inputs
        fig = plt.figure(figsize=(14, 7))
        for i in range(8):
            sample_heatmaps = heatmaps[i]
            ## Create a single summed heatmap for 
            joined_sample_heatmaps = sample_heatmaps.sum(dim=0)
            joined_sample_color_heatmap = cm(joined_sample_heatmaps)

            ax = fig.add_subplot(2, 4, i + 1, xticks=[], yticks=[])
            plt.imshow(joined_sample_color_heatmap)
        break
```

# Training Data Images

```python
display_images(TrainDataLoaderNoShuffle)
display_heatmaps(TrainDataLoaderNoShuffle)
```

# Validation Data Images 

```python
display_images(ValidationDataLoader)
display_heatmaps(ValidationDataLoader)
```

# Testing Data Images

```python
display_images(TestDataLoader)
display_heatmaps(TestDataLoader)
```
